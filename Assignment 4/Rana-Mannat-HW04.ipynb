```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,10,100)
y2=x**2
y3=x**3
```

```python
fig = plt.figure(1, figsize=[8, 8])

ax1 = fig.add_subplot(2,2,1)
ax1.plot(x, y2, 'r--', linewidth=4, label=r"$y = x^2$")
ax1.set_xlabel('x', fontsize=20)
ax1.set_ylabel('y', fontsize=20)
ax1.tick_params(axis='y', colors='red')
ax1.tick_params(axis='x', labelsize=15)
xticks = np.arange(0, 11, 5)
ax1.set_xticks(xticks)
ax1.set_xlim(0, 10)
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax1.legend(loc=2)
ax1.text(0, 110, '(a)', color='red', fontsize=20)
ax1.set_title(r"$y = x^2$", color='red', fontsize=20)

ax2 = fig.add_subplot(2,2,2)
ax2.plot(x, y3, 'b-.', linewidth=2, label=r"$y = x^3$")
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_xlabel('x', fontsize=20)
xticks = np.arange(0, 11, 5)
ax2.set_xticks(xticks)
ax2.set_ylabel('y', fontsize=20)
ax2.tick_params(axis='y', colors='blue', labelsize=20)
ax2.tick_params(axis='x', labelsize=15)
ax2.set_xlim(0, 10)
ax2.legend(loc=2)
ax2.text(0, 1100, '(b)', color='blue', fontsize=20)
ax2.set_title(r"$y = x^3$", color='blue', fontsize=20)

ax3 = fig.add_subplot(2,1,2)
ax3.plot(x, y2, 'r')
ax3.set_xlabel('x', fontsize=20)
ax3.set_ylabel('y', fontsize=20)
ax3.tick_params(axis='y', colors='red')
ax3.tick_params(axis='x', labelsize=20)
ax3.set_xlim(0, 10)
ax3.set_ylim(0, 100)
ax3.grid(color='r', alpha=0.1)
ax3.text(0, 90, '(c)', fontsize=20)
ax3.text(5, 40, r"$y = x^2$", color='red', fontsize=20)

ax4 = ax3.twinx()
ax4.plot(x, y3, 'b', label=r"$y = x^3$")
ax4.set_ylabel('y', fontsize=20)
ax4.set_ylim(0, 1000)
ax4.tick_params(axis='y', colors='blue', labelsize=20)
ax4.tick_params(axis='x', labelsize=20)
xticks = np.arange(0, 11, 1)
ax4.set_xticks(xticks)
ax4.set_xticklabels(["%.1f" % x for x in xticks])
ax4.set_xlim(0, 10)
ax4.text(6.5, 200, r"$y = x^3$", color='blue', fontsize=20)

fig.savefig('rana.pdf')
fig.tight_layout()
```
